```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
d=pd.read_csv("kc_house_data.csv")
```

```python
d.keys()
```

```python
x=d[["sqft_living"]]
x1=x
y=d["price"]
y1=y
```

```python
plt.scatter(x1,y1)
```

```python
i=[('polynomial',PolynomialFeatures(degree=15)),('model',LinearRegression())]
p=Pipeline(i)
```

```python
p.fit(x1,y1)
```

```python
reg=LinearRegression()
reg.fit(x1,y1)
yp=reg.predict(x1)
print(yp)
```

```python
pre=p.predict(x1)
plt.scatter(x1,y1,color="red")
plt.plot(x1,yp,color="blue",label="LinearRegression")
plt.plot(x1,pre,color="green",label="Polynomail Regression")
plt.show()
```

```python
print(sm.r2_score(y1,pre))
print(sm.r2_score(y1,yp))
```
